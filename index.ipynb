#### Curiosidade
Quando usamos loc estamos selecionando uma serie e aplicando uma condição de obtenção de dados. Por isso, se formos fazer filtros em Series não precisamos passá-lo como parâmetro:

````python
idades = [12, 13, 53, 56, 78, 21, 55, 34, 23, 23, 43]

serie_idades = pd.Series(idades)

#display(serie_idades) ## Mostrando todos as idades
display(serie_idades[serie_idades < 18]) ## Apenas menores de idade
````

Veja agora em um DataFrame a diferença:

````python
# Criando um DataFrame de exemplo com duas séries: idades e alturas
dados = {
    'idades': [12, 13, 53, 56, 78, 21, 55, 34, 23, 23, 43],
    'alturas': [1.60, 1.70, 1.80, 1.75, 1.85, 1.68, 1.73, 1.79, 1.65, 1.72, 1.78]
}

df = pd.DataFrame(dados)

# Usando loc para filtrar com base em uma condição
filtrado = df.loc[df['idades'] < 18]

display(filtrado)
````

## Adicionando e atualizando Colunas
> Exemplo: Digamos que a empresa vai fornecer um bonûs salarial para colaboradores com 10 ou mais anos de experiência. Para isso nosso analista de dados vai precisar criar uma nova coluna chamada "bonus" com 0 de valor padrão para quem não se encaixa nessa classe de veteranos e de 20 para esses funcionários antigos (valores em %).

Existem 2 maneiras de criar uma nova coluna, a primeira utilizando uma função e a segunda usando loc, mas antes vamos ver como selecionar partes da tabela sabendo a diferença entre loc e iloc:

```display(data.loc[:, :3])```

A sintaxe data.loc[:, :3] está tentando selecionar todas as linhas e as primeiras quatro colunas do DataFrame data. No entanto, a maneira correta de selecionar as primeiras quatro colunas seria usar **data.iloc[:, :4]**.

O método iloc é usado para seleção baseada em índices numéricos, enquanto o método loc é usado para seleção baseada em rótulos de índices. Como estamos tentando selecionar com base em números de índice das colunas, devemos usar iloc em vez de loc.

````python
display(data.iloc[:, :3])
````

Agora que sabemos analisar criaremos 2 colunas, com bonus iguais mas com métodos diferentes.

````python
data['bonus #1'] = data['Anos de Experiência'].apply(lambda x: 20 if x >= 10 else 0)
display(data)
````

````python
data.loc[:, 'bonus #2'] = 0
data.loc[data['Anos de Experiência'] >= 10, 'bonus #2'] = 20
display(data)
````

Vemos que no primeiro método nós declaramos uma nova coluna 'bonus #1' contendo o valor obtido em função de uma condição a partir das linhas da coluna 'Anos de Experiência'. Já no segundo método nós simplesmente passamos o método loc com : para indicar em todas as linhas e 'bonus #2'para indicar uma coluna nesse caso nova, passando o valor 0 e depois atualizando com 20.

## Adicionando e atualizando linhas

Para adicionar uma linha a um DataFrame no pandas, você pode usar o método `append()`. Aqui está como você pode fazer isso:

```python
# Suponha que você tenha um novo registro representado como um dicionário
novo_registro = {'Nome': 'João', 'Idade': 30, 'Cargo': 'Analista', 'Email': 'joao@example.com'}

# Use o método append para adicionar o novo registro ao DataFrame
data = data.append(novo_registro, ignore_index=True)
```

Neste exemplo, estamos adicionando um novo registro representado pelo dicionário `novo_registro` ao DataFrame `data`. Ao definir `ignore_index=True`, o pandas irá redefinir os índices do DataFrame resultante para garantir que eles sejam únicos e sequenciais.

Certifique-se de que o novo registro tenha as mesmas colunas e na mesma ordem que o DataFrame existente para evitar erros ao adicionar a linha.
